```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../Versuchsdaten_roh/crows.csv', sep=',')

plt.figure(figsize=(8, 6))
sns.boxplot(x='Lauf', y='Anzahl_Kraehen', palette="Set3", data=df)
plt.title('Wo und wie viele Krähen haben Sie auf dem Weg zum Bus aufgescheucht?')
plt.xlabel('Lauf')
plt.xticks([0, 1, 2], ['Referenzlauf', 'Virtuelles Tablet', 'Tablet Attrappe'])
plt.ylabel('Anzahl Krähen')
plt.savefig('../Visualisierungen/crows.png',dpi = 300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../Versuchsdaten_roh/usability.csv', sep=',')

plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='number', y='Praeferenz', orient='h')
plt.title('Wie gut war die Bedienbarkeit des Tablets in Bezug auf das reale Tablet?')
plt.xlabel('Einschätzung')
plt.xticks([1, 2, 3, 4, 5], ['Schlecht', 'Eher schlechter', 'Gleich', 'Eher besser', 'Gut'])
plt.yticks([0, 1], ['Virtuelles Tablet', 'Tablet Attrappe'])
plt.ylabel('Variante')
plt.savefig('../Visualisierungen/bedienbarkeit.png',dpi = 300)
plt.show()
```
