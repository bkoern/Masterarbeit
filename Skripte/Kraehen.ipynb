```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fontsizeHeader = 22
fontsizeLabel = 20
fontsize = 18

df = pd.read_csv('../Versuchsdaten_roh/crows.csv', sep=',')

plt.figure(figsize=(8, 6))
sns.boxplot(x='Lauf', y='Anzahl_Kraehen', palette="Set3", data=df)
plt.title('Wie viele Krähen haben Sie auf dem Weg zum Bus aufgescheucht?', fontsize=15)
plt.xlabel('Versuchsgruppen', fontsize=fontsizeLabel)
plt.xticks([0, 1, 2], ['Ohne', 'Virtuell', 'Haptisch'],fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.ylabel('Anzahl Krähen', fontsize=fontsizeLabel)
plt.savefig('../Visualisierungen/crows.png',dpi = 300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fontsizeHeader = 22
fontsizeLabel = 20
fontsize = 18

df = pd.read_csv('../Versuchsdaten_roh/usability.csv', sep=',')

#print(df)

plt.figure(figsize=(8, 6))
#sns.violinplot(data=df, x='number', y='Praeferenz', orient='h')
custom_palette = sns.color_palette("Set3", 8)[1:]
sns.countplot(data=df, x='number', hue='Tablet Variante', palette=custom_palette, edgecolor="black")

plt.title('Wie gut war die Bedienbarkeit des Tablets in Bezug auf das reale Tablet?', fontsize=15)
plt.xlabel('Einordung',fontsize=fontsizeLabel)
plt.xticks([0, 1, 2, 3, 4], ['Schlecht', 'Eher schlechter', 'Gleich', 'Eher besser', 'Gut'], fontsize=14)
#plt.yticks([0, 1], ['Virtuell', 'Attrappe'], fontsize=14)
plt.ylabel('Anzahl',fontsize=fontsizeLabel)
plt.xlim(-.5,4.5)
plt.savefig('../Visualisierungen/bedienbarkeit.png',dpi = 300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

def deutsch_format(x, pos):
    return str(x).replace('.', ',')

fontsizeHeader = 22
fontsizeLabel = 20
fontsize = 18

labels = ['Virtuell', 'Haptisch']
sizes = [21,8]

plt.figure(figsize=(6, 6))
custom_palette = sns.color_palette("Set3", 8)[1:]
wedges, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=0, colors=custom_palette)

for autotext in autotexts:
    autotext.set_text(autotext.get_text().replace('.', ','))

for wedge in wedges:
    wedge.set_edgecolor('black')

plt.axis('equal')

mpl.rcParams['font.size'] = 17

plt.title('Welche Variante hat Ihnen in Bezug auf die Realitätsnähe besser gefallen?',fontsize=11)

plt.savefig('../Visualisierungen/realitaetsnaehe.png',dpi = 300)
plt.show()
```
